```python
import pandas as pd
netflix_data = pd.read_csv("new_netflix_datas.csv")
netflix_data.head(5)
```

```python
user_data = pd.read_csv("fake_user_data.csv")
user_data.head(5)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
#Prepare the data for association learning by grouping the user-watched titles by username:
watched_titles = user_data.groupby('Username')['Watched Title'].apply(list).reset_index()
tr_data = list(watched_titles['Watched Title'].values)
```

```python
watched_titles.head(1000)
```

```python
# Apply TransactionEncoder to convert transaction list to encoded format
te = TransactionEncoder()
te_ary = te.fit(tr_data).transform(tr_data)
te_ary = te_ary.astype(int)
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
encoded_df.head()
```

```python
frequent_itemsets = apriori(encoded_df, min_support=0.005, use_colnames=True)
frequent_itemsets
```

```python
# Generate association rules from the frequent itemsets
association_results = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.01)
association_results
```

```python
def generate_recommendations(watching_title):
    # Find matching association rules for the user's selected show and sort by confidence
    matching_rules = association_results[association_results['antecedents'] == frozenset([watching_title])]
    matching_rules = matching_rules.sort_values('confidence', ascending=False)
     # Get recommended shows from the association rules
    recommended_shows = [item for sublist in matching_rules['consequents'].apply(list) for item in sublist]
    return recommended_shows[:3]
```

```python
import random
recommendations = generate_recommendations(random.choice(netflix_data.title))
recommendations
```
