```python
import pandas as pd
data = pd.read_csv("new_netflix_datas.csv")
data.head(5)
```

```python
from faker import Faker
fake = Faker()
num_users = 400
fake_user_data = []
```

```python
import random
from datetime import datetime, timedelta

for user_id in range(1, num_users + 1):
    num_watched = random.randint(25, 35) #each user should watch 25, 35 movies
    username = fake.user_name()
    country = fake.country()
    age = random.randint(15, 80)
    for _ in range(num_watched):
      title = random.choice(data['title'])
      date_watched = (datetime.now() - timedelta(days=random.randint(1, 1460))).strftime('%Y-%m-%d')
      percentage_watched = random.randint(1, 100)
      user_data = {
          'user_id': user_id,
          'username': username,
          'country': country,
          'user_age': age,
          'watched_title': title,
          'date_watched': date_watched,
          'percentage_watched': percentage_watched
      }
      fake_user_data.append(user_data)
```

```python
fake_user_data
```

```python
#saving in csv file
import csv
csv_file = 'fake_user_data.csv'
with open(csv_file, 'w', newline='') as file:
    # CSV yazıcıyı oluşturma
    writer = csv.writer(file)
    writer.writerow(['User ID', 'Username', 'Country', 'Watched Title', 'Age', 'Date Watched', 'Percentage Watched'])
    for user in fake_user_data:
        writer.writerow([
            user['user_id'],
            user['username'],
            user['country'],
            user['watched_title'],
            user['user_age'],
            user['date_watched'],
            user['percentage_watched']
        ])
```
